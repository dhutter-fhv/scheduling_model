```python
from visualization import util
import random
```

```python
# example data
example_solvers = ['Solver A', 'Solver B', 'Solver C', 'Solver D']
n_instances = 100
time_limit = 60
max_fitness = 100
```

```python
# generate sample data
data = []
for solver in example_solvers:
    solver_data = []
    fitness_data = []
    for i in range(n_instances):
        fitness = random.randint(1, max_fitness)
        fitness_data.append(fitness)
        # create fake progression data
        # technically, only improvements and their timestamps are required
        time_data = []
        for i in range(time_limit):
            timestamp = i
            time_data.append((timestamp, random.randint(fitness, max_fitness)))
        time_data.sort()
        solver_data.append((fitness, time_data))
    data.append(solver_data)
```

```python
overall_gaps = util.overall_gaps(data)
```

```python
util.ecdf_inf(overall_gaps, n_instances=n_instances, labels=example_solvers, xlabel='Gap', title='Fitness')
```

```python
util.ecdf_inf(overall_gaps, n_instances=n_instances, labels=example_solvers, xlim=(-0.1, 2.0))
```

```python
max_diff_fitness = 0.1

results_fitness_overall, results_fitness_self = util.get_all_timestamps(data, max_diff=max_diff_fitness)
```
