```python
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
def transform(x1, x2, sigma, mu):
    y1 = sigma * -np.sqrt(2*np.log(x1)) * np.cos(2 * np.pi * x2) + mu
    y2 = sigma * -np.sqrt(2*np.log(x1)) * np.sin(2 * np.pi * x2) + mu
    return y1, y2
```

```python
#y1 = sigma * sqrt(-2*ln x1) * cos (2 * pi * x2) + mu
#y2 = sigma * sqrt(-2*ln x1) * sin (2 * pi * x2) + mu
```

```python
original_values = []
transformed_values = []
sigma = 20
mu = 150.0
for i in range(1000):
    x = random.uniform(0, 100.0)
    y = random.uniform(0, 100.0)
    original_values.append((x, y))
    transformed_values.append(transform(x, y, sigma, mu))
```

```python
x_original = [x[0] for x in original_values]
y_original = [x[1] for x in original_values]
x_transformed = [x[0] for x in transformed_values]
y_transformed = [x[1] for x in transformed_values]

plt.scatter(x_original, y_original)
plt.scatter(x_transformed, y_transformed)
```

```python
def transform_jac(x, y, r, theta):
    return 1/2 *(x + y), 1/2 *(y-x)
```

```python
original_values = []
transformed_values = []
r = 20
theta = 150.0
for i in range(1000):
    x = random.uniform(0, 100.0)
    y = random.uniform(0, 100.0)
    original_values.append((x, y))
    transformed_values.append(transform_jac(x, y, r, theta))
```

```python
x_original = [x[0] for x in original_values]
y_original = [x[1] for x in original_values]
x_transformed = [x[0] for x in transformed_values]
y_transformed = [x[1] for x in transformed_values]

plt.scatter(x_original, y_original)
plt.scatter(x_transformed, y_transformed)
```
