```python
file_path = r'C:\Users\dhutt\Desktop\SCHEDULING_MODEL\code\reworked_data_model\results\tournament_testing2.txt'
file = open(file_path, 'r')
content = file.readlines()[1:]
```

```python
# 0 - source, 1 - instance, 2 - runtime, 3 - fevals, 4 - generations, 5 - restarts, 6 - fitness, 7 - adapt
tests = dict()
for line in content:
    values = line.split(';')
    key = f'{values[0]}{values[1]}'
    if key not in tests:
        tests[key] = []
    tests[key].append({'source': values[0], 'instance' : values[1], 'runtime': float(values[2]), 'fevals': int(values[3]), 'generations': int(values[4]), 'restarts': int(values[5]), 'fitness': float(values[6])})
```

```python
import statistics

data = dict()
for benchmark in tests:
    if benchmark not in data:
        data[benchmark] = dict()
    average_runtime = 0
    max_runtime = -float('inf')
    min_runtime = float('inf')
    std_runtime = []
    average_fevals = 0
    max_fevals = -float('inf')
    min_fevals = float('inf')
    std_fevals = []
    average_generations = 0
    max_generations = -float('inf')
    min_generations = float('inf')
    std_generations = []
    average_restarts = 0
    max_restarts = -float('inf')
    min_restarts = float('inf')
    std_restarts = []
    average_fitness = 0
    max_fitness = -float('inf')
    min_fitness = float('inf')
    std_fitness = []
    for test_run in tests[benchmark]:
        average_runtime += test_run['runtime']
        if test_run['runtime'] > max_runtime:
            max_runtime = test_run['runtime']
        if test_run['runtime'] < min_runtime:
            min_runtime = test_run['runtime']
        std_runtime.append(test_run['runtime'])
        average_fevals += test_run['fevals']
        if test_run['fevals'] > max_fevals:
            max_fevals = test_run['fevals']
        if test_run['fevals'] < min_fevals:
            min_fevals = test_run['fevals']
        std_fevals.append(test_run['fevals'])
        average_generations += test_run['generations']
        if test_run['generations'] > max_generations:
            max_generations = test_run['generations']
        if test_run['generations'] < min_generations:
            min_generations = test_run['generations']
        std_generations.append(test_run['generations'])
        average_restarts += test_run['restarts']
        if test_run['restarts'] > max_restarts:
            max_restarts = test_run['restarts']
        if test_run['restarts'] < min_restarts:
            min_restarts = test_run['restarts']
        std_restarts.append(test_run['restarts'])
        average_fitness += test_run['fitness']
        if test_run['fitness'] > max_fitness:
            max_fitness = test_run['fitness']
        if test_run['fitness'] < min_fitness:
            min_fitness = test_run['fitness']
        std_fitness.append(test_run['fitness'])
    
    data[benchmark]['average_runtime'] = average_runtime / len(tests[benchmark])
    data[benchmark]['max_runtime'] = max_runtime
    data[benchmark]['min_runtime'] = min_runtime
    data[benchmark]['std_runtime'] = statistics.stdev(std_runtime)

    data[benchmark]['average_fevals'] = average_fevals / len(tests[benchmark])
    data[benchmark]['max_fevals'] = max_fevals
    data[benchmark]['min_fevals'] = min_fevals
    data[benchmark]['std_fevals'] = statistics.stdev(std_fevals)

    data[benchmark]['average_generations'] = average_generations / len(tests[benchmark])
    data[benchmark]['max_generations'] = max_generations
    data[benchmark]['min_generations'] = min_generations
    data[benchmark]['std_generations'] = statistics.stdev(std_generations)

    data[benchmark]['average_restarts'] = average_restarts / len(tests[benchmark])
    data[benchmark]['max_restarts'] = max_restarts
    data[benchmark]['min_restarts'] = min_restarts
    data[benchmark]['std_restarts'] = statistics.stdev(std_restarts)

    data[benchmark]['average_fitness'] = average_fitness / len(tests[benchmark])
    data[benchmark]['max_fitness'] = max_fitness
    data[benchmark]['min_fitness'] = min_fitness
    data[benchmark]['std_fitness'] = statistics.stdev(std_fitness)

```
