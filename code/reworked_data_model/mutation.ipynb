```python
import math
import matplotlib.pyplot as plt
```

```python
x = 1 / 40
max_generations = 100
values = []
for i in range(1, max_generations):
    values.append(1 /( x + ((i * 1/max_generations) ** 4) * 0.5))
plt.plot([x for x in values])
```

```python
x = 1 / 40
max_generations = 25
values = []
for i in range(1, max_generations):
    values.append(math.exp(x + i * 1/max_generations))
plt.plot([x for x in values])
```

```python
x = 1 / 40
max_generations = 5
values = []
for i in range(1, max_generations):
    values.append(1 - math.log(x + i * 1/max_generations))
plt.plot([x for x in values])

```

```python
restarts = 10
max_generations = 5
values = []
ud = 0.4
pop_size = 5
pop_sizes = []
for i in range(0, restarts):
    values.append(max(1, int((ud * 0.2 * pop_size) + 0.5)))
    pop_sizes.append(pop_size)
    pop_size = min(pop_size * 2, 500)
plt.plot([x for x in values])
plt.plot([s for s in pop_sizes])
```

```python
def average_p(ap, t, t_last_improvement, t_restart, ap_max):
    return ap + (t - t_last_improvement * ((1 / (t_restart) ** 4))) * ap_max

ap = 1
aps = []
for i in range(10):
    aps.append(ap)
    ap = average_p(ap, i, 0, 10, 5)

plt.plot(aps)
```
